```python
from lazypredict.Supervised import LazyClassifier
```

```python
from datasets import load_dataset
from sklearn.model_selection import train_test_split
```

```python
from pathlib import Path
from glob import glob
```

```python
participants = [Path(path).stem for path in glob('./data/ecg_features/*.csv')]
train_participants, test_participants = train_test_split(participants, test_size=0.2)
train_participants, val_participants = train_test_split(train_participants, test_size=0.25)
```

```python
dataset = load_dataset(
    './data/ecg_features', 
    train_participants=train_participants,
    val_participants=val_participants,
    test_participants=test_participants
)
```

```python
X_labels = ['mean_hr', 'std_hr', 'HRV_TINN', 'hrv_index', 'nn50', 'HRV_pNN50', 'mean_hrv', 'std_hrv', 'rms_hrv', 'mean_fourier_frequencies', 'std_fourier_frequencies', 'sum_psd']
y_label = 'label'
```

```python
train = dataset['fit'].to_pandas()
test = dataset['test'].to_pandas()
```

# https://github.com/Edouard99/Stress_Detection_ECG/tree/main

```python
train = train.fillna(0)
test = test.fillna(0)
```

```python
X_train = train[X_labels]
y_train = train[y_label]

X_test = test[X_labels]
y_test = test[y_label]
```

```python
clf = LazyClassifier(verbose= 0, ignore_warnings= True, custom_metric= None)
```

```python
models, predictions = clf.fit(X_train, X_test, y_train, y_test)
models
```
