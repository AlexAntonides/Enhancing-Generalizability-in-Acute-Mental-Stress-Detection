```python
import pandas as pd
import pyarrow as pa
import glob
```

```python
import joblib
from joblib import Parallel, delayed
```

```python
def csv_to_parquet(file):
    df = pd.read_csv(
         file
    )
    df.to_parquet(file.replace('.csv', '.parquet').replace('ecg_preprocessed', 'parquet'))

Parallel(n_jobs=4)(delayed(csv_to_parquet)(file) for file in glob.glob('./data/ecg_preprocessed/*.csv'))
```
