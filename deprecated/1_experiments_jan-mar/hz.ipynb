```python
import neurokit2 as nk
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
signal = nk.ecg_simulate(duration=5, sampling_rate=1000, noise=10)

fft = np.fft.fft(signal)
fftfreq = np.fft.fftfreq(len(signal), 1/1000)

plt.ylabel("Amplitude")
plt.xlabel("Frequency [Hz]")
plt.ylim(0, 1100)
plt.plot(fftfreq,fft)
plt.show()

plt.figure(figsize=(15,6))
plt.plot(signal)
```
